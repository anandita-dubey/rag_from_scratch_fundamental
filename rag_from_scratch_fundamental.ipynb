```python
import shutil
import os

persist_dir = "/content/chroma_db"
shutil.rmtree(persist_dir, ignore_errors=True)  # delete old DB
os.makedirs(persist_dir, exist_ok=True)        # recreate folder
```

```python
#load the webpage you want to work with
from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader(
    web_paths=[
        # Saviynt official pages
        "https://knowledge.saviynt.com/",
        "https://saviynt.com/solution-guides/identity-governance-and-administration/",
        "https://saviynt.com/api-reference",
        "https://saviynt.com/university",
        "https://saviynt.com/exchange",
        "https://saviynt.com/resources-page",
        "https://saviynt.com/exchange/apps/documo",

        # Example Educosys page
        "https://www.educosys.com/course/genai"
    ]
)
docs = loader.load()
print(docs)
```

```python
#splitting docs
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
)
splits = text_splitter.split_documents(docs)
#print(splits)
#print(splits[1])
print(len(splits))
```

```python
#Adding docs to vectorDB
from langchain.embeddings.openai import OpenAIEmbeddings # To basically convert our docs(texts) into vectors to be stored in vectordb (chroma)
from langchain.vectorstores import Chroma #For storing the vector embeddings

vectorstore = Chroma.from_documents(documents=splits, embedding=OpenAIEmbeddings(), persist_directory="/content/chroma_db")
vectorstore.persist()
```

```python
#RAG Pipeline
retriever = vectorstore.as_retriever()
```

```python
#Augmentation
from langchain import hub
prompt = hub.pull("rlm/rag-prompt")
```

```python
#Setup LLM
from langchain_openai import ChatOpenAI
llm = ChatOpenAI()
```

```python
def format_docs(docs):
  return "\n".join(doc.page_content for doc in docs)
```

```python
#RagChain
from langchain_core.runnables import RunnablePassthrough, RunnableMap
from langchain_core.output_parsers import StrOutputParser

rag_chain = (
    RunnableMap({
        "context": retriever | format_docs,
        "question": RunnablePassthrough()
    })
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
rag_chain.invoke("What is Exam overview and link to access it")
```
